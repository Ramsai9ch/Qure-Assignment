```python
index=8153
for i in fips_state:
  s_dates=state.loc[state['fips']==i,'date'].values.tolist()
  s_c_dates=state_county.loc[state_county['state_fips']==i,'date'].unique().tolist()
  dates_to_add=[]
  for date in s_c_dates:
    if date not in s_dates:
      dates_to_add.append(date)
  print(len(dates_to_add),i)
  for j in dates_to_add:
      deaths=state_county.loc[(state_county['date']==j) & (state_county['state_fips']==i),"deaths"].values.tolist()
      index=index+1
      new_row = {'index': index,'date': j, 'state':dic[i], 'fips':i, 'cases':0, 'deaths':sum(deaths)}
      state= state.append(new_row, ignore_index=True)
```
