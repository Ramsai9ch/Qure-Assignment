```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

try:
    conn=mysql.connector.connect(host='localhost',user="root",password="Ramsai@sql",database="qure")
    cur=conn.cursor()
    conn.commit()
    for row,rs in state_county_data.iterrows():
        county_fips=rs[0]
        county_name=rs[1]
        state_name=rs[2]
        state_fips=rs[3]
        date=str(rs[4])
        deaths=rs[5]
        data_user=(county_fips, county_name, state_name, state_fips, date, deaths)
        insert_query = """INSERT INTO countyleveldata
                        (county_fips, county_name, state_name, state_fips, date, deaths) VALUES (%s, %s, %s, %s, '%s', %s)"""
        #print(insert_query)
                        
        cur.execute(insert_query,data_user)
    print("done")
    conn.commit()
    print(cur.rowcount, "Record inserted successfully into Laptop table")
    cur.close()
except Error as e:
    print(e)
```

```python
import mysql.connector as mysql

db = mysql.connect(
    host = "localhost",
    user = "root",
    passwd = "Ramsai@sql",
    database = "qure"
)
cursor=db.cursor()
cursor.execute("CREATE TABLE mobilitydata (state VARCHAR(255),county VARCHAR(255),date DATE, retail FLOAT, grocery_and_pharmacy FLOAT, parks FLOAT, transit_stations FLOAT, workplaces FLOAT, residential FLOAT)")
```

```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

try:
    conn=mysql.connector.connect(host='localhost',user="root",password="Ramsai@sql",database="qure")
    cur=conn.cursor()
    conn.commit()
    for row,rs in mobility_data.iterrows():
        state=rs[0]
        county=rs[1]
        date=str(rs[2])
        retail=rs[3]
        grocery_and_pharmacy=str(rs[4])
        parks=rs[5]
        transit_stations=rs[6]
        workplaces=rs[7]
        residential=rs[8]
        data_user=(state,county,date, retail, grocery_and_pharmacy, parks, transit_stations, workplaces, residential)
        insert_query = """INSERT INTO mobilitydata
                        (state,county, date, retail, grocery_and_pharmacy, parks, transit_stations, workplaces, residential) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"""
        #print(insert_query)
                        
        cur.execute(insert_query,data_user)
    print("done")
    conn.commit()
    print(cur.rowcount, "Record inserted successfully into Laptop table")
    cur.close()
except Error as e:
    print(e)
```
